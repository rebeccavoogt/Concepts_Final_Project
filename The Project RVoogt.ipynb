```python
# set up specials and menu dictionaries with items and their prices
specials = {"zebra" : 1099, "cheetah" : 3049, "warthog" : 1049}
menu = {"aardvark" : 499, "baboon" : 879, "cheetah" : 3999, "dolphin" : 899, "elephant" : 9999, 
       "fox" : 89, "goat" : 5, "hippo" : 499, "iguana": 499, "jaguar" : 769, "kangaroo" : 549, "lion" : 1299, "manatee" : 2999, 
        "owl" : 349, "peacock" : 349, "quail" : 89, "rhino" : 3699, "giraffe" : 789, "tiger" : 2579, 
        "polar bear" : 1699, "grizzly bear" : 1489, "bald eagle" : 5049,  "otter" : 899, "red panda" : 2349, 
       "seal" : 419, "tortoise" : 279, "vulture" : 549, "warthog" : 1229, "yak" : 1699, "zebra" : 1599}
# set total to 0 initially
total = 0
# make an empty dictionary to put ordered items into
runningorder = dict()
# make empty dictionaries for manager mode use
newitemdict = dict()
existdict = dict()

# start the program with a printed welcome
print("Hello, and welcome to the Animal Mart. The weekly specials are listed below.")
# display the specials in a nice format
print("\n".join("{}\t{}".format(k, v) for k, v in specials.items()))

# prompt the user for something they wish to order, change to lowercase and store it in the order variable, 
# can enter 'manager' to go into manager mode
order = input("You may type 'menu' at any time to see the menu with prices.\nWhat would you like to order? ").lower()

while True:
    # manager mode part of the loop
    if order == "manager":
        # ask user what menu they wish to order, change to lowercase, and store it in the variable 'question'
        question = input("Are you editing the menu or specials? ").lower()
        # if the user typed 'menu' then follow the loop below
        if question == "menu":
            # ask the user if they are adding, deleting, or changing items on the menu, change to lowercase, and store it in the variable 'prompt'
            prompt = input("Are you adding (a), deleting (d), or changing (c)? Type one of the letters to continue. ").lower()
            # if the user selected adding an item, follow the loop below
            if prompt == "a":
                # ask the user what the new item is, change to lowercase, and store it in the variable 'newkey'
                newkey = input("What is the new item? ").lower()
                # use the 'newkey' as the key in the empty dictionary 'newitemdict' and ask the user for the price, which
                # is then cast as an integer and set as the value to the key 'newkey'
                newitemdict[newkey] = int(input("What is the price in a round dollar amount? "))
                # print the new item and price
                print("\n\n".join("{}\t{}".format(k, v) for k, v in newitemdict.items()))
                # ask user if this item and price are correct, change to lowercase, and store in variable 'feedback'
                feedback = input("Is this correct? Type yes or no: ").lower()
                if feedback == "yes" or "y":
                    # if user confirms, update the menu with the new item and price
                    menu.update(newitemdict)
                    # delete the new item and price in case user is adding more than one item
                    del newitemdict[newkey]
                else:
                    # if user does not confirm, empty the dictionary of the item and price
                    del newitemdict[newkey]
            # else if the user selects deleting an item, follow the loop below
            elif prompt == "d":
                # if user wishes to delete an item, ask the user for the item, change to lowercase, and store it in variable 'delkey'
                delkey = input("What is the item to delete? ").lower()
                # print the item and its price from the menu dictionary for confirmation
                print(delkey + " " + str(menu[delkey]))
                # ask user for confirmation, change to lowercase, and store in variable 'feedback'
                feedback = input("Is this the item you wish to delete? Type yes or no: ").lower()
                if feedback == "yes" or "y":
                    # if user confirms, pop item out of menu, effectively deleting it and its price
                    menu.pop(delkey)
                else:
                    # if user does not confirm, do nothing
                    pass
            # else if the user selects changing an item, follow the loop below
            elif prompt == "c":
                # ask user for item to be changed, change to lowercase, and store in variable 'existkey'
                existkey = input("What item are you changing? ").lower()
                # ask user for the price, which is then cast as an integer and stored in the variable 'newprice'
                newprice = int(input("What is the price in a round dollar amount? "))
                # create a new entry in the empty dictionary 'existdict' with the user inputted 'existkey' as the key
                # and the 'newprice' as the value
                existdict[existkey] = newprice
                # print the item and price for user confirmation
                print("\n\n".join("{}\t{}".format(k, v) for k, v in existdict.items()))
                # ask user for confirmation, change to lowercase, and store in variable 'feedback'
                feedback = input("Is this correct? Type yes or no: ").lower()
                if feedback == "yes":
                    # if user confirms, use 'existdict' dictionary to update 'menu' dictionary
                    menu.update(existdict)
                    # delete item and key
                    del existdict[existkey]
                else:
                    # if user does not confirm, delete item and key in 'existdict'
                    del existdict[existkey]
            else:
                # if user enters something other than 'a','d', or 'c', continue the loop
                continue
            # after loop is complete, print 'Updated Menu:' with the 'menu' dict printed beneath
            print("Updated Menu:")
            print("\n".join("{}\t{}".format(k, v) for k, v in menu.items()))
        # else if the user typed 'specials', follow the loop below.
        elif question == "specials":
            # ask the user if they are adding, deleting, or changing items on the menu, change to lowercase, and store it in the variable 'prompt'
            prompt = input("Are you adding (a), deleting (d), or changing (c)? Type one of the letters to continue. ").lower()
            # if the user selects adding an item, follow the loop below
            if prompt == "a":
                # ask the user what the new item is, change to lowercase, and store it in the variable 'newkey'
                newkey = input("What is the new item? ").lower()
                # use the 'newkey' as the key in the empty dictionary 'newitemdict' and ask the user for the price, which
                # is then cast as an integer and set as the value to the key 'newkey'
                newitemdict[newkey] = int(input("What is the price in a round dollar amount? "))
                # print the new item and price
                print("\n\n".join("{}\t{}".format(k, v) for k, v in newitemdict.items()))
                # ask user if this item and price are correct, change to lowercase, and store in variable 'feedback'
                feedback = input("Is this correct? Type yes or no: ").lower()
                if feedback == "yes":
                     # if user confirms, update the specials menu with the new item and price
                    specials.update(newitemdict)
                    # delete the new item and price in case user is adding more than one item
                    del newitemdict[newkey]
                else:
                    # if user does not confirm, empty the dictionary of the item and price
                    del newitemdict[newkey]
            # else if the user selects deleting an item, follow the loop below
            elif prompt == "d":
                # if user wishes to delete an item, ask the user for the item, change to lowercase, and store it in variable 'delkey'
                delkey = input("What is the item to delete? ").lower()
                # print the item and its price from the specials dictionary for confirmation
                print(delkey + " " + str(specials[delkey]))
                 # ask user for confirmation, change to lowercase, and store in variable 'feedback'
                feedback = input("Is this the item you wish to delete? Type yes or no: ").lower()
                if feedback == "yes":
                    # if user confirms, pop item out of specials menu, effectively deleting it and its price
                    specials.pop(delkey)
                else:
                    # if user does not confirm, do nothing
                    pass
            # else if the user selects changing an item, follow the loop below
            elif prompt == "c":
                # ask user for item to be changed, change to lowercase, and store in variable 'existkey'
                existkey = input("What item are you changing? ").lower()
                # ask user for the price, which is then cast as an integer and stored in the variable 'newprice'
                newprice = int(input("What is the price in a round dollar amount? "))
                # create a new entry in the empty dictionary 'existdict' with the user inputted 'existkey' as the key
                # and the 'newprice' as the value
                existdict[existkey] = newprice
                # print the item and price for user confirmation
                print("\n\n".join("{}\t{}".format(k, v) for k, v in existdict.items()))
                # ask user for confirmation, change to lowercase, and store in variable 'feedback'
                feedback = input("Is this correct? Type yes or no: ").lower()
                if feedback == "yes":
                    # if user confirms, use 'existdict' dictionary to update 'specials' dictionary
                    specials.update(existdict)
                    # delete item and key
                    del existdict[existkey]
                else:
                    # if user does not confirm, delete item and key in 'existdict'
                    del existdict[existkey]
            else:
                # if user enters something other than 'a','d', or 'c', continue the loop
                continue
            # after loop is complete, print 'Updated Specials:' with the 'specials' dict printed beneath
            print("Updated Specials:")
            print("\n".join("{}\t{}".format(k, v) for k, v in specials.items()))
        else:
            # if user enters something other than 'menu' or 'specials' prompt the user for their order
            order = input("You may type 'menu' at any time to see the menu with prices.\nWhat would you like to order? ").lower()
            # use this value for order and go up a level in the loop
            continue
        # at the end of manager mode, prompt the user for their order
        order = input("You may type 'menu' at any time to see the menu with prices.\nWhat would you like to order? ").lower()
    # else if the user asks for their total, follow the loop below
    elif order == "total":
        # print a new line, and then the dictionary 'runningorder' which has their ordered items and their prices
        print("\n")
        print("\n".join("{}\t{}".format(k, v) for k, v in runningorder.items()))
        # print 'Subtotal     $' with the value of the variable total, which has the running total
        print("\nSubtotal      $" + str(total))
        # tell the user what their total with tax would be
        print("\nYour running total with tax is ${0:.2f}.".format(total*1.3))
        # prompt the user for their order
        order = input("\nIs there anything else you would like to order? \nIf you would like to see your running total, please enter 'total'. \nIf you would like to exit the ordering system and see your total, please enter 'done'." ).lower()
    # else if the user asks for the menu, follow the loop below
    elif order == "menu":
        # print the contents of the dictionary 'menu' with prices, nicely formatted
        print("\n".join("{}\t{}".format(k, v) for k, v in menu.items()))
        # prompt the user for their order
        order = input("\nIs there anything else you would like to order? \nIf you would like to see your running total, please enter 'total'. \nIf you would like to exit the ordering system and see your total, please enter 'done'." ).lower()
    # else if the user-input is in the 'specials' dictionary, follow the loop below
    elif order in specials:
        # add the value of the ordered item from the specials menu to the 'total' variable
        total = total + specials[order]
        # if the order isn't in the 'runningorder' dictionary, add it
        if order not in runningorder:        
            runningorder[order] = specials[order]
        # if the order is in the 'runningorder' dictionary, add a space to the key and add it to the 'runningorder' dictionary
        else:
            order2 = order + " "
            runningorder[order2] = specials[order]
        # prompt the user for their order
        order = input("\nIs there anything else you would like to order? \nIf you would like to see your running total, please enter 'total'. \nIf you would like to exit the ordering system and see your total, please enter 'done'." ).lower()
    # else if the user-input is in the 'menu' dictionary, follow the loop below
    elif order in menu:
        # add the value of the ordered item from the menu to the 'total' variable
        total = total + menu[order]
        # if the order isn't in the 'runningorder' dictionary, add it
        if order not in runningorder:        
            runningorder[order] = menu[order]
            # if the order is in the 'runningorder' dictionary, add a space to the key and add it to the 'runningorder' dictionary
        else:
            order2 = order + " "
            runningorder[order2] = menu[order]
        # prompt the user for their order
        order = input("\nIs there anything else you would like to order? \nIf you would like to see your running total, please enter 'total'. \nIf you would like to exit the ordering system and see your total, please enter 'done'." ).lower()
    # else if the user inputted 'done', follow the loop below
    elif order == "done":    
        # print a new line, then print the items ordered and prices from the 'runningorder' dictionary
        print("\n")
        print("\n".join("{}\t{}".format(k, v) for k, v in runningorder.items()))
        # print 'Subtotal     $' with the value of the 'total' variable
        print("\nSubtotal      $" + str(total))
        # tell the user what their total with tax is
        print("\nYour total bill with sales tax is ${0:.2f}. Thank you for your purchase.".format(total*1.3))
        # exit the loop
        break
    else:
        # if the user-input doesn't match 'manager', 'total', 'done', or an item in the 'menu' or 'specials' dictionaries
        # tell them they ordered something we don't have
        print("You have ordered something that we don't have. Sorry! Please order something on the menu or specials list. ")
        # prompt the user for their order
        order = input("\nIs there anything else you would like to order? \nIf you would like to see your running total, please enter 'total'. \nIf you would like to exit the ordering system and see your total, please enter 'done'." ).lower()
        # continue the loop with the new value for the 'order' variable
        continue    
```
